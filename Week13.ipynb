```python
import csv
import matplotlib.pyplot as plt


with open('seoul.csv') as file:
    data=csv.reader(file)
    header = next(data)
    high, low = [], []
    for row in data:
        if row[1] != '':
            if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '30':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
    plt.plot(high,label='max_temp')
    plt.plot(low,label='min_temp')
    plt.legend()
    plt.show()
```

```python
with open('seoul.csv') as file:
    data=csv.reader(file)
    header = next(data)
    january, august = [], []
    
    for row in data:
        if row[4] != '':
            if row[0].split('-')[1] == '01':
                january.append(float(row[4]))
            elif row[0].split('-')[1] == '08':
                august.append(float(row[4]))
    
    plt.hist(august, bins=100, color='red', label='Aug')
    plt.hist(january, bins=100, color='blue', label='Jan')
    plt.legend()
    plt.show()
```
