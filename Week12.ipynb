```python
## 실습 1 서울의 일교차

import csv


max_gap_day, max_temp_day, min_temp_day = '', '', ''
max_gap, max_temp, min_temp = 0, -999, 999

with open("seoul.csv", encoding='cp949') as f:
    data = csv.reader(f, delimiter=",")
    header = next(data)

    for row in data:
        if max_gap < float(row[4]) - float(row[3]):
            max_gap = float(row[4]) - float(row[3])
            max_gap_day = row[0]
        
        if max_temp < float(row[4]):
            max_temp = float(row[4])
            max_temp_day = row[0]

        if min_temp > float(row[3]):
            min_temp = float(row[3])
            min_temp_day = row[0]


print("일교차가 가장 큰 날")
print(max_gap_day)
print("최저기온이 가장 낮은 날")
print(min_temp_day)
print("최고기온이 가장 높은 날")
print(max_temp_day)
```

```python
import csv

with open("subwayfee.csv") as f:
    data = csv.reader(f, delimiter=",")
    header = next(data)

    subway_lines = {}

    for row in data:
        temp = subway_lines.get(row[1])
        sum_passenger = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
        station = row[3]
        if temp:
            temp[station] = sum_passenger
        else:
            subway_lines[row[1]] = {station: sum_passenger}

    for key in subway_lines.keys():
        print(key)
        line = sorted(subway_lines[key].items(), key=lambda x:x[1], reverse=True)

        print("\n이용객이 가장 많은 역 순위")
        for i in range(1, 4):
            print(f"{i}. {line[i][0]}({line[i][1]})")
        print("\n이용객이 가장 적은 역 순위")
        for i in range(1, 4):
            print(f"{i}. {line[-i][0]}({line[-i][1]})")
        print("==================\n")
```
